# Q5: Go to webpage https://www.amazon.in/ Enter “Laptop” in the search field and then click the search icon. Then set CPU
Type filter to “Intel Core i7” as shown in the below image:

Aftersetting the filters scrape first 10 laptops data. You have to scrape 3 attributes for each laptop:
1. Title
2. Ratings
3. Price

```python
from bs4 import BeautifulSoup
```

```python
driver4 = webdriver.Chrome()
driver4.get("https://www.amazon.in/")
```

```python
title=[]
rating=[]
price=[]
```

```python
prod1 = driver4.find_element(By.XPATH, '//div[@class="nav-search-field "]/input')
prod1.send_keys('Laptop')
```

```python
submit1 = driver4.find_element(By.XPATH, '//div[@class="nav-right"]/div/span/input')
submit1.click()
```

```python
def extract_record(item):
    """Extract and return the data from a single record."""
    atag = item.h2.a
    description = atag.text.strip()
    
    try:
        # Price
        price_parent = item.find('span', 'a-price')
        price = price_parent.find('span', 'a-offscreen').text
    except AttributeError: 
        price = ''
    
    try:
        # Rating
        rating = item.i.text
    except AttributeError:
        rating = ''
        
    result = (description, price, rating)
    return result

records = []
soup = BeautifulSoup(driver4.page_source, 'html.parser')
results = soup.find_all('div', {'data-component-type': 's-search-result'})

for item in results:
    record = extract_record(item)
    if record:
        records.append(record)
        
records = records[:10]

titles = []
ratings = []
prices = []

for record in records:
    titles.append(record[0])
    prices.append(record[1])
    ratings.append(record[2])

intel_core_i7_amazon = pd.DataFrame({"Title": titles, "Ratings": ratings, "Price": prices})
intel_core_i7_amazon
```

# Q6: Write a python program to scrape data for Top 1000 Quotes of All Time. The above task will be done in following steps:
1. First get the webpagehttps://www.azquotes.com/
2. Click on TopQuote
3. Than scrap
4. a)Quote b) Author c) Type Of Quotes

```python
driver4 = webdriver.Chrome()
driver4.get("https://www.azquotes.com/")
```

```python
quote=[]
auth=[]
typ=[]
```

```python
top = driver4.find_element(By.XPATH, '//a[@href="/top_quotes.html"]')

top.click()
```

```python
quotes_collected = 0
while quotes_collected < 1000:

    quotes_on_page = driver4.find_elements(By.XPATH, '//a[@class="title"]')
    authors_on_page = driver4.find_elements(By.XPATH, '//div[@class="author"]/a')
    types_on_page = driver4.find_elements(By.XPATH, '//div[@class="tags"]')

    for i in range(min(len(quotes_on_page), len(authors_on_page), len(types_on_page))):
        quote.append(quotes_on_page[i].text)
        auth.append(authors_on_page[i].text)
        typ.append(types_on_page[i].text)
        quotes_collected += 1

        if quotes_collected >= 1000:
            break

    if quotes_collected >= 1000:
        break

    next_button = driver4.find_element(By.CLASS_NAME, "next")

    next_button.click()
```

```python
az=pd.DataFrame({'Quote':quote,'Author':auth,'Type':typ})
az
```
